## Read file

```python
def readfiles(pathtofile, file_name, ext_name,withoutneutral,withoutnan,minq,minfert,logfert,line_to_begin):
    data_file = pathtofile+file_name+"."+ext_name
    data = dict()
    params = []
    nbline=1
    with io.open(data_file,'r') as f:
        reader = csv.reader(f)
        firstline = True
        for line in reader:
            lin = line[0].split()
            if not firstline:
                if nbline>=line_to_begin:
                    for j in range(len(lin)):
                        if ext_name == "allele":
                            if float(lin[5])>minq and float(lin[6])>minfert:
                                if withoutneutral==1 and float(lin[1]) != -3 :
                                    if withoutnan == 0:
                                        if logfert==True and (j==6 or j==9):
                                            data[params[j]].append(np.log(float(lin[j])))
                                        else:
                                            data[params[j]].append(float(lin[j]))
                                    elif withoutnan ==1 :
                                        if lin[5]!="-nan" and lin[6]!="-nan":
                                            if logfert==True and (j==6 or j==9):
                                                data[params[j]].append(np.log(float(lin[j])))
                                            else:
                                                data[params[j]].append(float(lin[j])) 
                                elif withoutneutral==0:
                                    if withoutnan == 0:
                                        if logfert==True and (j==6 or j==9):
                                            data[params[j]].append(np.log(float(lin[j])))
                                        else:
                                            data[params[j]].append(float(lin[j]))
                                    elif withoutnan ==1 :
                                        if lin[5]!="-nan" and lin[6]!="-nan":
                                            if logfert==True and (j==6 or j==9):
                                                data[params[j]].append(np.log(float(lin[j])))
                                            else:
                                                data[params[j]].append(float(lin[j]))
                        elif ext_name == "trace" and logfert==True and (j==6 or j==14):
                            data[params[j]].append(np.log(float(lin[j])))
                        else :
                            data[params[j]].append(float(lin[j]))
                nbline=nbline+1
            else:
                for j in lin:
                    data[j]=[]
                    params.append(j)
                firstline = False
                nbline=nbline+1
    return (data)
```

## Functions for plot

```python
def graph_per_allele(target_folder_path, file_name, ext_name, x_axis_param, y_axis_param,withoutneutral,withoutnan,minq,minfert,logfert,line_to_begin):
    data_all = readfiles(target_folder_path, file_name, 'allele',withoutneutral,withoutnan,minq,minfert,logfert,line_to_begin)
    data_trace = readfiles(target_folder_path, file_name, 'trace',withoutneutral,withoutnan,minq,minfert,logfert,line_to_begin)
    x_per_allele=dict()
    y_per_allele=dict()
    for i in range(len(data_all["Generation_number"])):
        cle = data_all['Allele_number'][i]
        if ((str(cle) in x_per_allele) == False and cle!=-3):
            x_per_allele[str(cle)]=[]
            x_per_allele[str(cle)].append(data_all[x_axis_param][i])
            y_per_allele[str(cle)]=[]
            if y_axis_param=="selection":
                #print(data_all["Generation_number"][i]/100)
                sel_coef=np.log(data_all['fertility_analytic'][i])-np.log(data_trace['fertility_analytic'][int(data_all["Generation_number"][i]/100)])
                y_per_allele[str(cle)].append(sel_coef)
            else:
                y_per_allele[str(cle)].append(data_all[y_axis_param][i])
        elif ((str(cle) in x_per_allele) == True and cle!=-3):
            x_per_allele[str(cle)].append(data_all[x_axis_param][i])
            if y_axis_param=="selection":
                #print(data_all["Generation_number"][i]/100)
                sel_coef=np.log(data_all['fertility_analytic'][i])-np.log(data_trace['fertility_analytic'][int(data_all["Generation_number"][i]/100)])
                y_per_allele[str(cle)].append(sel_coef)
            else:
                y_per_allele[str(cle)].append(data_all[y_axis_param][i])
    return(x_per_allele,y_per_allele)
```

```python
boldallel=-1
def plot_regim_per_allele(target_folder_path, file_name, ext_name, x_param, y_params, withoutneutral, withoutnan,minq,minfert,logfert,line_to_begin, allelenbmin, allelenbmax,boldallel,xmin,xmax):
    index_line=1
    #legend_list=["Frequency","Activity","mean_affinity","q_allele","Fertility_allele","q_analytic","fertility_analytic"]
    #legend_list=["Frequency of\n PRDM9 alleles","Proportion of\n active sites","mean_affinity","q_allele","Fertility_allele","q_analytic","fertility_analytic"]
    #legend_list=["Frequence des\n allèles PRDM9","Proportion des\n cibles actives","Affinite moyenne","q_allele","Fertilite des \n allèles PRDM9","q_analytic","Fertilite des \n allèles PRDM9"]
    #legend_list=["Frequence des\n allèles PRDM9","Proportion des\n cibles actives","Affinite moyenne","Fertilite des \n allèles PRDM9"]
    #legend_list=["Frequence des\n allèles PRDM9","Coefficient de\n selection"]
    legend_list=["Frequency of\n PRDM9 alleles","Proportion of\n active sites","Selection \n Coefficient"]
    #legend_list=["freq","act","fert"]
    lettres=["A","B","C","D","E"]
    #lettres=["B"]
    figure=plt.figure(figsize=(10,7))
    #figure=plt.figure(figsize=(20,10))
    for i in y_params:
        plt.subplot(len(y_params),1,index_line) 
        (X,Y)=graph_per_allele(target_folder_path, file_name, ext_name, x_param, i, withoutneutral,withoutnan,minq,minfert,logfert,line_to_begin)
        #print(X)
        for j in range(allelenbmax-allelenbmin):
            allele_nb=allelenbmin+j
            nameallelelist=[str(allele_nb),'.0']
            nameallele=''.join(nameallelelist)
            if X.get(nameallele, 0)!=0:
                #plt.plot(X[nameallele], Y[nameallele], '.')
                #print(nameallele)
                if(float(nameallele)==boldallel):
                    #print("a")
                    plt.plot(X[nameallele], Y[nameallele],linewidth=6)
                else:
                    plt.plot(X[nameallele], Y[nameallele])
        plt.ylabel(legend_list[index_line-1],fontsize=16)
        #plt.xlabel("Number of generations") 
        #plt.text(x=-7800,y=0.55,s=lettres[index_line-1],fontweight='bold',fontsize=16)
        if index_line==len(y_params):
            plt.xlabel("Number of generations",fontsize=16)
            #plt.xlabel("Nombre de generations",fontsize=20)
        #plt.title("With genic dosage")
        plt.xticks(fontsize=14)
        plt.yticks(fontsize=14)
        if i=="selection":
            plt.axhline(y=0, color='black')
        plt.xlim(xmin,xmax)
        index_line+=1
```

**Caption: Evolution of the summary statitsitcs of the model for each allele along time (nmber of generations) for the monomorphic regime** (N = 5000, u=5e-6, v=5e-5)

In this regime, at each time, only one allele dominate the population at a frequency near 1, until the invasion of a new allele that will take its place very quickly.

In details : an allele appears in the population and invade it to reach a frequency near 1. At the same time, its activity (the proportion of still activated target site for this allele) begins to decrease with the erosion due to mutations and biased gene conversion. When a certain state of erosion is reached by the allele, when a new one appears in the population, it is somehow positively selected and replace the ancient allele. It then erodes its targets until beeing replaced and so on. This cyclic process is called the Red Queen process.
